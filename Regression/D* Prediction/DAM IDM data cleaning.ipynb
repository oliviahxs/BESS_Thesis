```python
#convert into average DAM price in each EFA block

# fill the EFA block column with the correct values
import pandas as pd

# Read the cleaned data
df = pd.read_csv('Cleaned DAM IDM price.csv')

# Convert Time (UK) to datetime
df['Time (UK)'] = pd.to_datetime(df['Time (UK)'], format='%d/%m/%Y %H:%M')

# Function to assign EFA block based on hour
def assign_efa_block(hour):
    if hour >= 23 or hour < 3:
        return 1
    elif 3 <= hour < 7:
        return 2
    elif 7 <= hour < 11:
        return 3
    elif 11 <= hour < 15:
        return 4
    elif 15 <= hour < 19:
        return 5
    else:  # 19 <= hour < 23
        return 6

# Assign EFA block numbers
df['EFA block'] = df['Time (UK)'].dt.hour.apply(assign_efa_block)

# Create a date column for grouping (for blocks that cross midnight)
df['date'] = df['Time (UK)'].dt.date

# Adjust date for EFA block 1 (23:00-03:00) - midnight crossing
# For hours 23:00, we want them grouped with the next day's early morning hours
df['efa_date'] = df['date']
df.loc[df['Time (UK)'].dt.hour == 23, 'efa_date'] = df.loc[df['Time (UK)'].dt.hour == 23, 'date'] + pd.Timedelta(days=1)

# Group by adjusted date and EFA block, then calculate average DAM price
efa_averages = df.groupby(['efa_date', 'EFA block'])['DAM'].mean().reset_index()
efa_averages.columns = ['efa_date', 'EFA block', 'avg_dam_price']

# Merge back to original dataframe
df = df.merge(efa_averages, on=['efa_date', 'EFA block'], how='left')
df['Average DAM price in EFA block'] = df['avg_dam_price']

# Clean up temporary columns
df = df.drop(columns=['date', 'efa_date', 'avg_dam_price'])

# Convert Time (UK) back to string format
df['Time (UK)'] = df['Time (UK)'].dt.strftime('%d/%m/%Y %H:%M')

# Save the updated data
df.to_csv('Cleaned DAM IDM price.csv', index=False)

print("EFA block assignment and averaging completed!")
print(f"Total rows: {len(df)}")
print("\nFirst 24 rows with EFA blocks:")
print(df[['Time (UK)', 'DAM', 'EFA block', 'Average DAM price in EFA block']].head(24))
print("\nLast 24 rows with EFA blocks:")
print(df[['Time (UK)', 'DAM', 'EFA block', 'Average DAM price in EFA block']].tail(24))
```

```python
import pandas as pd

# Read the cleaned data
df = pd.read_csv('Cleaned DAM IDM price.csv')
df['Time (UK)'] = pd.to_datetime(df['Time (UK)'], format='%d/%m/%Y %H:%M')

# Create delivery date (for EFA block 1, use next day)
df['delivery_date'] = df['Time (UK)'].dt.date
df.loc[df['EFA block'] == 1, 'delivery_date'] = df.loc[df['EFA block'] == 1, 'delivery_date'] + pd.Timedelta(days=1)

# Group by delivery date and EFA block to get single average per block
efa_summary = df.groupby(['delivery_date', 'EFA block'])['DAM'].mean().reset_index()
efa_summary.columns = ['Delivery Date', 'EFA Block', 'Average DAM Price']

# Format delivery date
efa_summary['Delivery Date'] = pd.to_datetime(efa_summary['Delivery Date']).dt.strftime('%d/%m/%Y')

# Save result
efa_summary.to_csv('EFA_Block_DAM_Averages.csv', index=False)

print(f"Created {len(efa_summary)} EFA block averages")
print("\nFirst 12 EFA blocks:")
print(efa_summary.head(12))
```

```python
# Minimal code for EFA block aggregation
import pandas as pd

# Read the data
df = pd.read_csv('Cleaned DAM IDM price.csv')
df['Time (UK)'] = pd.to_datetime(df['Time (UK)'], format='%d/%m/%Y %H:%M')
df['Hour'] = df['Time (UK)'].dt.hour

# Assign EFA blocks and delivery dates
def assign_efa_block(hour):
    if hour >= 23: return 1
    elif hour < 3: return 1
    elif hour < 7: return 2
    elif hour < 11: return 3
    elif hour < 15: return 4
    elif hour < 19: return 5
    else: return 6

df['EFA_Block'] = df['Hour'].apply(assign_efa_block)

# Adjust delivery date for EFA block 1 (which spans midnight)
df['Delivery_Date'] = df['Time (UK)'].dt.date
df.loc[df['Hour'] >= 23, 'Delivery_Date'] = (df.loc[df['Hour'] >= 23, 'Time (UK)'] + pd.Timedelta(days=1)).dt.date

# Calculate EFA block averages (one row per block)
efa_block_averages = df.groupby(['Delivery_Date', 'EFA_Block'])['DAM'].mean().reset_index()
efa_block_averages.rename(columns={'DAM': 'Average_DAM_Price'}, inplace=True)

# Save and display
efa_block_averages.to_csv('EFA_Block_DAM_Averages.csv', index=False)
print(f"Created {len(efa_block_averages)} EFA block averages")
print(efa_block_averages.head(12))  # Show first 2 days
```
